#### Goal

The goal of this assignment is to estimate the position $ \mathbf{p}_t $ of an agent, for $t = {1, ..., \text{T}}$ The agent moves in an indoor environment and gets information about its approximate change in position $ \Delta \mathbf{p}_t $ at every time-step. To model this, you may assume a linear dynamic model of the form:
$$\mathbf{p}_{t + 1} = \mathbf{p}_t + \Delta \mathbf{p}_{t+1} + \mathbf{w}_{t+1}$$

Where:

- $\mathbf{p}_t$ is a 3-dimensional position vector of the agent at time $t$. 

- $\Delta \mathbf{p}_t$ is the change in positon.

- $\mathbf{w}_t \sim \mathcal{N}(0_{3 \times 1}, \mathbf{Q})$ is Gaussian white noise.

- $ \mathbf{Q} \in \mathbb{R}^{3 \times 3} $ is the covariance of the process noise.

Since the information of the change in position $ \Delta \mathbf{p}_t$ contains errors, the position estimates from the dynamic model will drift. This means that over time, the agent will lose track of where it actually is. To remedy this, the agent is carrying a magnetometer, and has a map of the magnetic field. The magnetometer samples a measurement of the magnetic field $\mathbf{y}_t$ at every position $\mathbf{p}_t$. The map of the magnetic field is modeled by a non-linear function: a reduced rank Gaussian process (you do not need to understand, model or change the GP model, only to call the function evaluation). This results in the non-linear measurement model:
$$ \mathbf{y}_t = f(\mathbf{p}_t) + \mathbf{v}_t $$

Where:

- $\mathbf{y}_t$ is a 3-dimensional measurement vector of the magnetic field at time-step $t$.

- $f(\cdot)$ is the reduced rank Gaussian process function.

- $\mathbf{v}_t \sim \mathcal{N}(0_{3 \times 1}, \mathbf{R})$ is Gaussian white noise.

- $ \mathbf{R} \in \mathbb{R}^{3 \times 3} $ is the covariance of the measurement noise.

### Particle Filter Implementation

One way to reduce the positional drift observed in the figure above, is to use magnetometer measurements. To this end, you will design a particle filter. Your task is to fill in the following three functions.

#### Task
Fill in the `dynamicUpdate()` function according to the dynamic model given above. 

#### Note:
You do not have to change the inputs and outputs to the given functions.


```python

def dynamicUpdate(particlePositions, 
                  deltaPosition,
                  processCovariance,
                  Nparticles):
    
    ''' Dynamic update '''
    # goal is to generate particles for step k+1 using assumed dynamic model
    # based on particle filter function, particles generated by this function are used in meas. update
    # 
    # inputs: particle positions from step k, delta positions, covariance of gaussina process noise, num of particles
    # output: new particle positions

    for i in range(Nparticles):
        particlePositions[:,[i]] = particlePositions[:,[i]] + deltaPosition + np.random.multivariate_normal(mean=np.zeros(3), cov=processCovariance).reshape(3,1) # dynamic model
    return particlePositions
```

#### Discuss

The dynamicUpdate function is used to reflect the system's expected update based on a known model of the system. It updates the particles positions to align with the natural evolution of the true system state in time. 

#### Task

Fill in the `measurementUpdate()` according to the measurement model given above. 

To implement the measurement model, you will need to make predictions of the magnetic field. You DO NOT need to create this function yourself. To make predictions, use the function `GP.makeMagneticFieldPrediction(positions, modelParameters)` function make sure the input `positions` $\in \mathbb{R}^{3 \times \text{T}} $. The output should be a matrix $\in \mathbb{R}^{3 \times \text{T}}$ magnetic field predictions for every input location.

```python
def measurementUpdate(particlePositions, 
                      particleWeights,
                      magnetometerMeasurement,
                      measurementCovariance,
                      Nparticles,
                      modelParameters):
    
    ''' Measurement update '''
    predictionsMagneticFieldParticles = GP.makeMagneticFieldPrediction(particlePositions, modelParameters)

    # likelihood of particles given measurement
    #error = np.ones((1,Nparticles)) @ magnetometerMeasurement - predictionsMagneticFieldParticles
    error = np.repeat(magnetometerMeasurement, Nparticles, axis=1) - predictionsMagneticFieldParticles

    for i in range(0,Nparticles-1):
        # refer to slide 24
        particleWeights[i] = np.exp(-0.5 * error[:,i].T @ np.linalg.inv(measurementCovariance) @ error[:,i]) * particleWeights[i] # I assume p(y_k|x_k) / unsure
    
    # normalize weights
    # commented out because it is done in resampling step
    # particleWeights = particleWeights / np.sum(particleWeights)
    return particleWeights
```

#### Discuss
The measurement update is the key filtering step in the particle-based approach. It finds the likelihood of each particle given all of the information available and assigns it a weight accordingly. Those particles with a high likelihood are then used to inform our filtered system where the low likelihood particles are ignored / have minimal impact. 

#### Task

Fill in the remainder of the `particleFilter()` function. 


```python
def particleFilter(particlePositions, 
                   deltaPosition, 
                   magnetometerMeasurement, 
                   processCovariance, 
                   measurementCovariance, 
                   Nparticles,
                   modelParameters):
    ''' Estimates position based on a known map'''

    particleWeights = 1/Nparticles * np.ones((Nparticles,1)) # vector of weights for each particle

    if modelParameters['NtimeStepPF'] > 0:
        ''' Dynamic update '''
        particlePositions = dynamicUpdate(particlePositions, 
                                          deltaPosition, 
                                          processCovariance, 
                                          Nparticles)  


    ''' Measurement update '''
    (particleWeights) = measurementUpdate(particlePositions, 
                                             particleWeights,
                                             magnetometerMeasurement, 
                                             measurementCovariance, 
                                             Nparticles,
                                             modelParameters)


    meanPosition = np.reshape(np.mean(particlePositions, axis=1), (3,1))
    positionNorms = np.linalg.norm(particlePositions-np.repeat(meanPosition, Nparticles, axis=1), axis = 0)
    #maxPositionID = np.argmax(positionNorms)
    #minPositionID = np.argmin(positionNorms)
    #maxPosition = particlePositions[:,[minPositionID]] # using max is surely the opposite of best
    indexMax = np.argmax(particleWeights)
    maxPosition = particlePositions[:,[indexMax]] # Select the particle with the highest weight
    

    ''' Resampling '''    
    if np.sum(particleWeights) > 1e-12:
        particleWeights /= np.sum(particleWeights)
        indices = helper.systematicResample(particleWeights)
        particlePositions = particlePositions[:, indices]
    else:
        print('Degenerate particles: All the weights are close to zero !!!')
        particleWeights /= np.sum(particleWeights)
        indices = helper.systematicResample(particleWeights)


    return particlePositions, meanPosition, maxPosition
```

#### Discuss

Particle resampling is the process by which unlikely particles are removed from consideration and new particles are generated which match better with available source information. There is still some level of randomness involved in selecting particles for removal (i.e. no fixed barrier for removal at X% of particles), with particles that are chosen to be resampled adopting the position of one that'll remain (one of the more likely particles).

Comment on the reason for choosing a particle filter for this estimation problem.

Particle filters are used in cases in which the probability distribution of the unknown variable according to measurements follow some non-gaussian or multimodal distribution. In the case above, we want to combine a drifting dynamic model and the magnetic field measurements to obtain a more accurate position estimate. Both sources of information are subject to gaussian noise, but they will inevitably have peaks in different locations. The resulting distribution can be characterized as multimodal, warranting the use of a particle filter for position estimation.
